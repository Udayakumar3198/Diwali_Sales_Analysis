# Exploratory Data Analysis

```python
gender=sns.countplot(x="Gender",data=df,hue="Gender", palette={'M': 'skyblue', 'F': 'salmon'},legend=False)
for bars in gender.containers:
    gender.bar_label(bars)    
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Count Plot of Gender')
plt.show()
```

```python
Gen_sales=df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
Gen_sales
```

```python
gen=sns.barplot(x="Gender",y="Amount",data=Gen_sales,hue="Gender", palette={'M': 'yellow', 'F': 'lightgreen'},legend=False)
for bars in gen.containers:
    gen.bar_label(bars) 
plt.xlabel('Gender')
plt.ylabel('Sales')
plt.title('Bar Plot of Gender wise Sales')
plt.show()
```

 * From the above charts, we can observe that Females purchased more products than Males

```python
Age=sns.countplot(x="Age Group",data=df,hue="Gender", palette="tab20",legend=True)
for bars in Age.containers:
    Age.bar_label(bars)    
plt.xlabel('Age_Gender')
plt.ylabel('Count')
plt.title('Count Plot of Age_Gender')
plt.show()
```

* From the above graph, we can observe that most of the buyers are from the age_gender 26-25 and in that females purcharsed high products than males.

```python
States=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(5)
States
```

```python
state_amount=sns.barplot(x="State",y="Amount",data=States,palette="tab20",hue="State",legend=False)
for bars in state_amount.containers:
    state_amount.bar_label(bars)    
plt.xlabel('States')
plt.ylabel('Amount')
plt.title('State wise amount')
plt.show()
```

* From the above graph , we can observe that "Uttar Pradesh" purchased more products

```python
State=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(5)
State
```

```python
state_order=sns.barplot(x="State",y="Orders",data=State,hue="State",legend=False)
for bars in state_order.containers:
    state_order.bar_label(bars)    
plt.xlabel('States')
plt.ylabel('Orders')
plt.title('State wise order')
plt.show()
```

* From the above graph , we can observe that "Uttar Pradesh" has ordered more products which has the count of 4807 orders. Maharashtra and Karnataka are in the next two places.

```python
Marital_status=sns.countplot(x="Marital_Status",data=df,hue="Marital_Status",legend =False)
sns.set(rc={"figure.figsize":(6,5)})
for bars in Marital_status.containers:
    Marital_status.bar_label(bars)
plt.title("Count of Marital_status") 
plt.show()
    
```

```python
a=df.groupby(['Marital_Status','Gender'],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
a
```

```python
M_G=sns.barplot(x="Marital_Status",y="Amount",data=a,hue="Gender")
sns.set(rc={"figure.figsize":(6,5)})
for bars in M_G.containers:
    M_G.bar_label(bars)
plt.title("Amount spent based on Marital_status and gender") 
plt.show()
    
```

*  From the above graph,we can observe that unmarried females spent more amount on purchasing product.
